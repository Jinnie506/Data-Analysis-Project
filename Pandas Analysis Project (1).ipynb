## Step 4 - Unique Values in Columns

With the line of code below, we can get a list of the categorical column in the dataset. We iterate through the columns and check if each element is an object datatype. The result is saved to the "categorical_columns" variable
as a list.

```python
#Uncomment the code and Run it
categorical_columns = [col for col in company_df.columns if company_df[col].dtype == "object" ]
categorical_columns
```

#### To-Do- From the information above, generate the unique values in the categorical columns (apart from the example - Branch column).

```python
#Example - Uncomment the Code and Run it
company_df['Branch'].unique().tolist()
```

```python
print("Total Number of unique values in the Branch Column : {}". format(len(company_df['Branch'].unique().tolist())))
```

```python
for colm in categorical_columns:
    if colm != 'Branch':
        print("Total Number of unique values in the {} Column : {}". format(colm, len(company_df[colm].unique().tolist()))) 
```

#### To-Do: The `value_counts()` function is used to get a Series containing counts of unique values. For the categorical columns above, generate the count figure of the values using the `value_counts() ` method.


```python
for colm in categorical_columns:
    print('Count of unique values in the {} Column :'.format(colm))
    print(company_df[colm].value_counts())
    print('\n')
```

#### To-Do - Using the groupby object, display a table that shows the gross income of each city, and determine the city with the highest total gross income.

#### To-Do -  An interesting insight to explore is the interaction of Unit price on the Quantity of goods purchased. To achieve this: 
* Use the `catplot()` to plot `Product line per unit price`, and `Product line per Quantity`. Set the kind parameter to `point` 
* In a new cell, Write a summary of the insights you uncovered

```python
sns.catplot(x = 'Product line', y = 'Unit price', data = company_df, kind = 'point', aspect = 4)
plt.title('Product line per unit price')
```

* The products with the highest unit price is Fashion accessories and also Sports and travel, then we have Food and beverages with Home and lifestyle having almost equal unit price, then Health and beauty while Electronic accessories has the lowest unit price.

```python
sns.catplot(y='Quantity', x='Product line', data=company_df, kind='point', aspect=4)
plt.title('Product line per quantity')
```

* Fashion accessories cost more and looks to be purchased less by customers.
* While Electronic accessories cost less and is purchased more by customers.
* In general, products that have lower cost are purchased more by customers across the three branches.
